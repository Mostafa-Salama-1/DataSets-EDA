```python
q1=df.duration_ms.quantile(0.25)
q2=df.duration_ms.quantile(0.75)
IQR=q2-q1
upper=q2+(1.5*IQR)
lower=q1-(1.5*IQR)
outliears=df.loc[((df.duration_ms>upper)|(df.duration_ms< lower))]
len(outliears)
```

```python
new_df=df.copy()
new_df.loc[((df.duration_ms>upper),"duration_ms")]=upper
new_df.loc[((df.duration_ms<lower),"duration_ms")]=lower
len(new_df)
```

```python
new_df = new_df.drop_duplicates()
new_df.shape
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

# Fit the encoder and transform the 'track_genre' column
new_df['track_genre'] = label_encoder.fit_transform(new_df['track_genre'])
new_df
```

```python
y=new_df['track_genre'].copy()
new_df=new_df.drop(['track_genre'],axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

# Fit the scaler on the data and transform it
df_scaled = scaler.fit_transform(new_df)

# Convert back to a DataFrame if needed
df_scaled = pd.DataFrame(df_scaled, columns=new_df.columns)
new_df=df_scaled
```

```python
new_df
```

```python
fig, axs = plt.subplots(4, 2, figsize=(12, 16))

# List of columns you want to plot
columns_to_plot = ['popularity', 'duration_ms', 'danceability', 'energy', 'key', 'mode', 'speechiness', 'acousticness']

# Iterate over the axes and columns to create each plot
for i, ax in enumerate(axs.flat):
    # Plot a histogram for each column in the grid
    ax.hist(new_df[columns_to_plot[i]], bins=30, edgecolor='black')
    ax.set_title(f'Histogram of {columns_to_plot[i]}')
    ax.set_xlabel(columns_to_plot[i])
    ax.set_ylabel('Frequency')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import seaborn as sns
corr_matrix = new_df.corr()
plt.figure(figsize=(12, 10))

# Plot the heatmap
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True, linewidths=0.5)
#sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True, linewidths=0.5, annot_kws={"size": 10})

# Set the title
plt.title('Correlation Heatmap of Spotify Dataset')

# Display the heatmap
plt.show()
```
